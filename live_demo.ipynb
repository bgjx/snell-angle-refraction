# LIVE DEMO :
### Python Code to calculate take-off angle, travel-time, and incidence angle at station

```python
import refraction as refract
model_top = [
                    [-3.00, -1.90], [-1.90, -0.59], [-0.59, 0.22], [0.22, 2.50], [2.50, 7.00], [7.00, 9.00], [9.00, 15.00], [15.00, 33.00], [33.00, 99999.00] 
    ]
velocity_P = [2.68, 2.99, 3.95, 4.50, 4.99, 5.60, 5.80, 6.40, 8.00]

hypo = [  37.916973, 126.651613, 200]   # lat, long, depth (negative down)
station = [  37.916973, 126.700882, 2200]     # lat, long, elevation (positive up)
```

### > if you want to produce the report and plot the image

```python
refract.calculate_inc_angle(hypo, station, model_top, velocity_P, plot_figure = True)                            
# function call returns the plot image and tuple of the take-off angle, travel time, and the station's incidence angle.
```

### > if you want to return the function call as an object without producing the plot image

```python
take_off, travel_time, incidence = refract.calculate_inc_angle(hypo, station, model_top, velocity_P, plot_figure = False)                            
# function call only returns the tuple of the take-off angle, travel time, and the station's incidence angle.

print(f"take of angle: {take_off}")
print(f"total travel time: {travel_time}")
print(f"incidence angle at station: {incidence}")
```
